```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
test_data = pd.read_csv("test.csv")
train_data = pd.read_csv("train.csv")
```

```python
y_train = train_data["Survived"]
```

```python
y_train.head()
```

```python
train_data.drop("Survived", axis=1, inplace=True)
```

```python
train_data.head()
```

```python
train_data.info()
```

```python
def red_table(table):
    table.drop(["Name", "Ticket", "Cabin", "Embarked"], axis=1, inplace=True)
    table["Sex"] = table["Sex"].map({"male": 0, "female": 1})
    table["Age"].fillna(table["Age"].median(), inplace=True)
    table["Fare"].fillna(table["Fare"].median(), inplace=True)

    return table
```

```python
train_data = red_table(train_data)
```

```python
train_data.head()
```

```python
test_data = red_table(test_data)
```

```python
test_data.head()
```

# Дерево решений

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
tree_params = {"max_depth": list(range(1, 11)), "min_samples_leaf": list(range(1, 10))}

tree = GridSearchCV(DecisionTreeClassifier(), tree_params, cv=10)

tree.fit(train_data, y_train)
```

```python
print("Best params:", tree.best_params_)
print("Best cross validaton score", tree.best_score_)
```

```python
tree_predictions = tree.predict(test_data[train_data.columns])
```

```python
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': tree_predictions})
output.to_csv('submission.csv', index=False)
```

Решение: 73,44%

# Случайный лес

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
rf = RandomForestClassifier(n_estimators=100)
rf.fit(train_data, y_train)

param_grid = {
    'n_estimators': [50, 100, 200],       
    'max_depth': [None, 10, 20, 30],      
    'min_samples_split': [2, 5, 10],      
    'min_samples_leaf': [1, 2, 4],        
    'bootstrap': [True, False]           
}


grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5)

grid_search.fit(train_data, y_train)
```

```python
print("Best params:", grid_search.best_params_)
print("Best cross validaton score", grid_search.best_score_)
```

```python
tree_predictions = grid_search.predict(test_data[train_data.columns])
```

```python
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': tree_predictions})
output.to_csv('submission.csv', index=False)
```

Решение: 76,31%

# Логистическая регрессия

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
%%time
scaler = StandardScaler()
train_data_scaled = scaler.fit_transform(train_data)
model = LogisticRegression(random_state=17, max_iter=1000)
param_grid = {
    'penalty': ['l2'],  
    'C': [0.1, 1.0],
    'solver': ['liblinear', 'lbfgs'],  
    'max_iter': [100, 500, 1000],  
    'tol': [1e-4]  
}
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, n_jobs=-1)
grid_search.fit(train_data_scaled, y_train)
```

```python
print("Лучшие параметры:", grid_search.best_params_)
print("Лучший результат на кросс-валидации:", grid_search.best_score_)
```

```python
train_data_scaled = scaler.fit_transform(train_data)  # масштабирование
train_data_scaled = pd.DataFrame(train_data_scaled, columns=train_data.columns)  # сохраняем имена признаков

model.fit(train_data_scaled, y_train)

test_data_scaled = scaler.transform(test_data)
test_data_scaled = pd.DataFrame(test_data_scaled, columns=test_data.columns)  # сохраняем имена признаков

predictions = model.predict(test_data_scaled)
```

```python
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
```
